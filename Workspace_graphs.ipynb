```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('pose_workspaceXZ.csv')
df2 = pd.read_csv('manip_workspaceXZ.csv')

x = df.values[:,1]
y = df.values[:,2]
z = df.values[:,3]
v = df2.values[:,1]
```

```python
fig, ax = plt.subplots()
im = ax.scatter(x, z, c=v)
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('manipulability', rotation=270)
plt.ylabel("z[m] ")
plt.xlabel("x[m] ")
plt.savefig('workspace.jpg', format='jpg')
plt.show()
```
