```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('pose_workspace5.csv')
df2 = pd.read_csv('manip_workspace5.csv')
df3 = pd.read_csv('E1_workspace5.csv')

x = df.values[:,1]
y = df.values[:,2]
z = df.values[:,3]
m = df2.values[:,1]
v1 = df3.values[:,1]
v2 = df3.values[:,2]
v3 = df3.values[:,3]
v4 = df3.values[:,4]
v5 = df3.values[:,5]
v6 = df3.values[:,6]
```

```python
fig, ax = plt.subplots()
im = ax.scatter(x, z, c=m,vmin = 0, vmax=0.2)
cbar = plt.colorbar(im, ax=ax)
cbar.set_label('manipulability', rotation=270)
plt.ylabel("z[m] ")
plt.xlabel("x[m] ")
#plt.savefig('workspace.jpg', format='jpg')
plt.show()
```

```python
f = plt.figure()
#f, ax = plt.subplots()
f, ax = plt.subplots(nrows = 3, ncols=2, sharex=True, sharey=True)
sc = ax[0][0].scatter(x, z, c=v1, vmin = 0, vmax=1)
ax[0][0].set_xlabel("x[m]",labelpad=5)
ax[0][0].set_ylabel("z[m]",labelpad=5)

ax[0][1].scatter(x, z, c=v2, vmin = 0, vmax=1)
ax[0][1].set_xlabel("x[m]",labelpad=5)
ax[0][1].set_ylabel("z[m]",labelpad=5)

ax[1][0].scatter(x, z, c=v3, vmin = 0, vmax=1)
ax[1][0].set_xlabel("x[m]",labelpad=5)
ax[1][0].set_ylabel("z[m]",labelpad=5)

ax[1][1].scatter(x, z, c=v4, vmin = 0, vmax=1)
ax[1][1].set_xlabel("x[m]",labelpad=5)
ax[1][1].set_ylabel("z[m]",labelpad=5)

ax[2][0].scatter(x, z, c=v5, vmin = 0, vmax=1)
ax[2][0].set_xlabel("x[m]",labelpad=5)
ax[2][0].set_ylabel("z[m]",labelpad=5)

ax[2][1].scatter(x, z, c=v6, vmin = 0, vmax=1)
ax[2][1].set_xlabel("x[m]",labelpad=5)
ax[2][1].set_ylabel("z[m]",labelpad=5)
f.tight_layout()
f.colorbar(sc, ax=ax)
plt.show()
```
